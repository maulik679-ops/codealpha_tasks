**Step 2: Load Data & Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Load the data
df = pd.read_csv("real_estate_clean.csv")

print(f"Data loaded! {len(df)} rows")
df.head()
```

**Step 3: Bar Chart — Average Price by City**

```python
# Average price per city
avg_price = df.groupby("City")["Price_Lakhs"].mean().round(2).sort_values(ascending=False)

# Plot
plt.figure(figsize=(8, 5))
plt.bar(avg_price.index, avg_price.values, color=["#FF6B6B", "#4ECDC4", "#45B7D1", "#96CEB4"])
plt.title("Average Property Price by City (in Lakhs)", fontsize=14)
plt.xlabel("City")
plt.ylabel("Average Price (Lakhs)")

# Add value labels on top of each bar
for i, val in enumerate(avg_price.values):
    plt.text(i, val + 5, f"₹{val}L", ha="center", fontsize=10)

plt.tight_layout()
plt.savefig("chart1_avg_price_by_city.png", dpi=150)
plt.show()
print("Chart saved!")
```

**Step 4: Bar Chart — Average Area by City**

```python
# Average area per city
avg_area = df.groupby("City")["Area_sqft"].mean().round(2).sort_values(ascending=False)

# Plot
plt.figure(figsize=(8, 5))
plt.bar(avg_area.index, avg_area.values, color=["#A8E6CF", "#FFD3B6", "#FFAAA5", "#FF8B94"])
plt.title("Average Property Area by City (in sqft)", fontsize=14)
plt.xlabel("City")
plt.ylabel("Average Area (sqft)")

# Add value labels on top of each bar
for i, val in enumerate(avg_area.values):
    plt.text(i, val + 5, f"{val} sqft", ha="center", fontsize=10)

plt.tight_layout()
plt.savefig("chart2_avg_area_by_city.png", dpi=150)
plt.show()
print("Chart saved!")
```

**Step 5: Pie Chart — Property Status Distribution**

```python
# How many properties are Ready to Move vs Under Construction
status_counts = df["Status"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(status_counts.values, labels=status_counts.index, autopct="%1.1f%%")
plt.title("Property Status Distribution")
plt.savefig("chart3_status_distribution.png", dpi=150)
plt.show()
```

**Step 6: Bar Chart — Price per Sqft by City**

```python
# Average price per sqft by city
avg_ppsf = df.groupby("City")["Price_per_sqft"].mean().round(2).sort_values(ascending=False)

plt.figure(figsize=(8, 5))
plt.bar(avg_ppsf.index, avg_ppsf.values, color="skyblue")
plt.title("Average Price per Sqft by City (₹)")
plt.xlabel("City")
plt.ylabel("Price per Sqft (₹)")
plt.tight_layout()
plt.savefig("chart4_price_per_sqft.png", dpi=150)
plt.show()
```

**Step 7: Scatter Plot — Price vs Area**

```python
# Does bigger area mean higher price?
plt.figure(figsize=(8, 5))
plt.scatter(df["Area_sqft"], df["Price_Lakhs"], alpha=0.6, color="coral")
plt.title("Price vs Area")
plt.xlabel("Area (sqft)")
plt.ylabel("Price (Lakhs)")
plt.tight_layout()
plt.savefig("chart5_price_vs_area.png", dpi=150)
plt.show()
```
