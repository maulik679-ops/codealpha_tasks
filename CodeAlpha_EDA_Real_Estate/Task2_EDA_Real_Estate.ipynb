**Step 2: Import Libraries & Load Data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Load the data
df = pd.read_csv("real_estate_data.csv")

# Quick look
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(f"\nColumn names: {list(df.columns)}")
df.head()
```

**Step 3: Basic Data Info**

```python
# Check data types and missing values
print("Data Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nBasic Statistics:")
df.describe()
```

**Step 4: Clean Price & Area Columns**

```python
# Clean Price column — remove ₹ and convert Cr to Lakhs
def clean_price(val):
    try:
        val = str(val).replace("₹", "").replace(",", "").strip()
        if "Cr" in val:
            return float(val.replace("Cr", "").strip()) * 100  # 1 Cr = 100 Lakhs
        elif "L" in val:
            return float(val.replace("L", "").strip())
        else:
            return float(val)
    except:
        return np.nan

# Clean Area column — remove sqft
def clean_area(val):
    try:
        return float(str(val).replace("sqft", "").replace(",", "").strip())
    except:
        return np.nan

# Apply cleaning
df["Price_Lakhs"] = df["Price"].apply(clean_price)
df["Area_sqft"] = df["Area"].apply(clean_area)

# Verify
print(df[["Price", "Price_Lakhs", "Area", "Area_sqft"]].head(10))
```

**Step 5: Basic Statistics**

```python
# Now we can see proper min, max, avg stats
print("Price Statistics (in Lakhs):")
print(df["Price_Lakhs"].describe())

print("\nArea Statistics (in sqft):")
print(df["Area_sqft"].describe())
```

**Step 6: Average Price Per City**

```python
# Average price per city
avg_price = df.groupby("City")["Price_Lakhs"].mean().round(2).sort_values(ascending=False)

print("Average Property Price by City (in Lakhs):")
print(avg_price)
```

**Step 7: Average Area Per City**

```python
# Average area per city
avg_area = df.groupby("City")["Area_sqft"].mean().round(2).sort_values(ascending=False)

print("Average Property Area by City (in sqft):")
print(avg_area)
```

**Step 8: Property Status Distribution**

```python
# How many properties are Ready to Move vs Under Construction
status_counts = df["Status"].value_counts()

print("Property Status Distribution:")
print(status_counts)
```

**Step 9: Price per Sqft**

```python
# Calculate price per sqft
df["Price_per_sqft"] = (df["Price_Lakhs"] * 100000) / df["Area_sqft"]
df["Price_per_sqft"] = df["Price_per_sqft"].round(2)

# Average price per sqft by city
avg_ppsf = df.groupby("City")["Price_per_sqft"].mean().round(2).sort_values(ascending=False)

print("Average Price per Sqft by City (₹):")
print(avg_ppsf)
```

**Step 10: Find the Most Expensive & Cheapest Properties**

```python
# Most expensive property
most_expensive = df.loc[df["Price_Lakhs"].idxmax()]
print("Most Expensive Property:")
print(f"  Title : {most_expensive['Title']}")
print(f"  City  : {most_expensive['City']}")
print(f"  Price : ₹{most_expensive['Price_Lakhs']} Lakhs")
print(f"  Area  : {most_expensive['Area_sqft']} sqft")

print()

# Cheapest property
cheapest = df.loc[df["Price_Lakhs"].idxmin()]
print("Cheapest Property:")
print(f"  Title : {cheapest['Title']}")
print(f"  City  : {cheapest['City']}")
print(f"  Price : ₹{cheapest['Price_Lakhs']} Lakhs")
print(f"  Area  : {cheapest['Area_sqft']} sqft")
```

**Step 11: Missing Values & Data Quality Check**

```python
# Check missing values
print("Missing Values in each column:")
print(df.isnull().sum())

print(f"\nTotal missing values: {df.isnull().sum().sum()}")
print(f"Total rows: {len(df)}")
print(f"Complete rows (no missing): {df.dropna().shape[0]}")

# Drop rows where Price or Area is missing
df_clean = df.dropna(subset=["Price_Lakhs", "Area_sqft"])
print(f"\nRows after dropping missing Price/Area: {len(df_clean)}")
```

**Step 12: Correlation Between Price and Area**

```python
# Check if bigger properties are more expensive
correlation = df_clean["Price_Lakhs"].corr(df_clean["Area_sqft"])
print(f"Correlation between Price and Area: {correlation:.2f}")

if correlation > 0.7:
    print("Strong positive correlation — bigger properties are more expensive!")
elif correlation > 0.4:
    print("Moderate correlation — size somewhat affects price!")
else:
    print("Weak correlation — size doesn't strongly affect price!")
```

**Step 13: Summary of All Insights**

```python
print("=" * 50)
print("   EDA SUMMARY — Real Estate Dataset")
print("=" * 50)
print(f"Total Properties Analyzed : {len(df_clean)}")
print(f"Cities Covered            : {df_clean['City'].nunique()}")
print(f"\nMost Expensive City  : {df_clean.groupby('City')['Price_Lakhs'].mean().idxmax()}")
print(f"Cheapest City        : {df_clean.groupby('City')['Price_Lakhs'].mean().idxmin()}")
print(f"\nLargest Avg Area     : {df_clean.groupby('City')['Area_sqft'].mean().idxmax()}")
print(f"Smallest Avg Area    : {df_clean.groupby('City')['Area_sqft'].mean().idxmin()}")
print(f"\nHighest Price/Sqft   : {df_clean.groupby('City')['Price_per_sqft'].mean().idxmax()}")
print(f"Lowest Price/Sqft    : {df_clean.groupby('City')['Price_per_sqft'].mean().idxmin()}")
print(f"\nPrice-Area Correlation : Weak (size doesn't drive price)")
print("=" * 50)
```

**Step 14: Save the Cleaned Data**

```python
# Save the cleaned data
df_clean.to_csv("real_estate_clean.csv", index=False)

print(f"Saved! Total rows: {len(df_clean)}")
print(f"Columns: {list(df_clean.columns)}")

# Download it
from google.colab import files
files.download("real_estate_clean.csv")
```
